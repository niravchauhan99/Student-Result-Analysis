```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df = pd.read_csv("Expanded_data_with_more_features.csv")
print(df.head())
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

# Drop Unnamed Column

```python
df = df.drop("Unnamed: 0", axis = 1)
```

```python
print(df.head())
```

# Gender Distribution

```python
plt.figure(figsize = (5, 5))
ax = sb.countplot(data = df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

# Above graph we have analyzed that:
# Number of females is more than that of males in the dataset

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean', "ReadingScore": 'mean', "WritingScore": 'mean'})
gb
```

```python
# plt.figure(figsize = (4, 4))
sb.heatmap(gb, annot = True)
plt.title("Parent's Education vs Student's Score")
plt.show()
```

# Based on the heatmap analysis, it appears that parental education significantly influences their child's academic outcomes.

```python
marital_gb = df.groupby("ParentMaritalStatus").agg({"MathScore": 'mean', "ReadingScore": 'mean', "WritingScore": 'mean'})
marital_gb
```

```python
sb.heatmap(marital_gb, annot = True)
plt.title("Parent's Marital Status vs Student's Score")
plt.show()
```

# For Parent's Marital Status, there is negligible impact in their child's result

# Ethnic Group Distribution

```python
# print(df["EthnicGroup"].unique())

groupA = df.loc[(df["EthnicGroup"] == "group A")].count()
groupB = df.loc[(df["EthnicGroup"] == "group B")].count()
groupC = df.loc[(df["EthnicGroup"] == "group C")].count()
groupD = df.loc[(df["EthnicGroup"] == "group D")].count()
groupE = df.loc[(df["EthnicGroup"] == "group E")].count()

l = ["group A", "group B", "group C", "group D", "group E"]
myList = [groupA["EthnicGroup"], groupB["EthnicGroup"], groupC["EthnicGroup"], groupD["EthnicGroup"], groupE["EthnicGroup"]]
print('myList is:', myList)

plt.pie(myList, labels = l, autopct = "%1.2f%%")
plt.title("Distribution of Ethnic Groups")
plt.show()
```
